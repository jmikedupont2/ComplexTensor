in torch complex numbers are supported throught complex tensor types allowing for operations on data with both real and imaginary components.

here is an example:

```python
# Define real and imaginary parts
real_part = torch.tensor([1.0, 2.0])
imaginary_part = torch.tensor([3.0, 4.0])

# Combine into a complex tensor
complex_tensor = torch.complex(real_part, imaginary_part)
print("Complex Tensor:", complex_tensor)
```

to create a complex tensor, you can combine real and imaginary parts:

`Complex Tensor: tensor([1.+3.j, 2.+4.j])`

you can access the real and imaginary components of a complex tensor using the .real and .imag attributes:


```python
print("Real Part:", complex_tensor.real)
print("Imaginary Part:", complex_tensor.imag)
```

torch supports various operations on complex tensors, such as addition, multiplication and matrix multiplication:

```python
# Create another complex tensor
another_complex_tensor = torch.complex(torch.tensor([5.0, 6.0]), torch.tensor([7.0, 8.0]))

# Addition
sum_tensor = complex_tensor + another_complex_tensor
print("Sum:", sum_tensor)

# Multiplication
product_tensor = complex_tensor * another_complex_tensor
print("Product:", product_tensor)
```

while torch supports complex tensors, not all neural network modules are fully compatible with complex data, for instance, certain activation functions and layers may not support complex inputs directly, however, there are third-party libraries, such as complexPyTorch, that provide complex-valued neural network components.




Here's an example of how you might define a complex-valued linear layer using the complexPyTorch library:

```python
# Install complexPyTorch
!pip install complexPyTorch

import torch
from complexPyTorch.complexLayers import ComplexLinear
from complexPyTorch.complexFunctions import complex_relu

# Define a complex-valued linear layer
class ComplexModel(torch.nn.Module):
    def __init__(self):
        super(ComplexModel, self).__init__()
        self.fc1 = ComplexLinear(2, 2)

    def forward(self, x):
        x = self.fc1(x)
        x = complex_relu(x)
        return x

# Instantiate the model
model = ComplexModel()

# Create a complex input tensor
input_tensor = torch.complex(torch.tensor([[1.0, 2.0]]), torch.tensor([[3.0, 4.0]]))

# Forward pass
output = model(input_tensor)
print("Output:", output)
```

```python
# Proof Of Entanglement
import torch
import numpy as np
import logging
from complextensor import ComplexTensor

# Set up logging
logging.basicConfig(level=logging.DEBUG)
logger = logging.getLogger(__name__)

class EntanglementProcessor:
    def __init__(self, n_qubits, device='cpu'):
        self.n_qubits = n_qubits
        self.state_size = 2 ** n_qubits
        self.device = device

    def create_bell_state(self):
        logger.debug("Creating Bell State.")
        # Prepare a Bell state generalized for an arbitrary number of qubits
        # |00...0⟩ + |11...1⟩ / sqrt(2)
        alpha = 1 / np.sqrt(2)
        dtype = torch.float64  # Ensure consistent dtype for real and imaginary parts
        bell_real = torch.zeros(self.state_size, dtype=dtype, device=self.device)
        bell_real[0] = alpha  # Initialize |00...0>
        bell_real[-1] = alpha  # Initialize |11...1>
        bell_imag = torch.zeros(self.state_size, dtype=dtype, device=self.device)
        return ComplexTensor(bell_real, bell_imag)

    def get_entanglement_entropy(self, state, partition):
        logger.debug(f"Calculating entanglement entropy for partition: {partition}")
        try:
            shape = [2] * self.n_qubits
            state_reshaped = state.forward().view(shape)
            rho_A = self._partial_trace(state_reshaped, partition)
            eigenvalues = torch.linalg.eigvalsh(rho_A)
            eigenvalues = eigenvalues[eigenvalues > 1e-10]  # Filter near-zero eigenvalues
            entropy = -torch.sum(eigenvalues * torch.log2(eigenvalues)).item()
            return entropy
        except Exception as e:
            logger.error(f"Error calculating entanglement entropy: {e}")
            raise

    def _partial_trace(self, state, partition):
        logger.debug("Performing partial trace.")
        try:
            total_qubits = len(state.shape)
            traced_dims = [i for i in range(partition, total_qubits)]
            rho = torch.tensordot(state, state.conj(), dims=(traced_dims, traced_dims))
            rho_shape = [2 ** partition, 2 ** partition]
            rho_A = rho.view(*rho_shape)
            return rho_A
        except Exception as e:
            logger.error(f"Error in partial trace calculation: {e}")
            raise

class SuperpositionProcessor:
    def __init__(self, device='cpu'):
        self.device = device

    def create_equal_superposition(self, n_qubits):
        logger.debug(f"Creating equal superposition for {n_qubits} qubits.")
        real_part = torch.full((2**n_qubits,), 1/np.sqrt(2**n_qubits), dtype=torch.float64, device=self.device)
        imag_part = torch.zeros(2**n_qubits, dtype=torch.float64, device=self.device)
        return ComplexTensor(real_part, imag_part)

    def apply_phase_shift(self, state, angle):
        logger.debug(f"Applying phase shift by angle: {angle}")
        try:
            angle_tensor = torch.tensor(angle, dtype=torch.float64, device=state.real.device)
            cos_angle = torch.cos(angle_tensor)
            sin_angle = torch.sin(angle_tensor)
            new_real = state.real * cos_angle - state.imag * sin_angle
            new_imag = state.real * sin_angle + state.imag * cos_angle
            return ComplexTensor(new_real, new_imag)
        except Exception as e:
            logger.error(f"Error applying phase shift: {e}")
            raise

# Testing Code
if __name__ == "__main__":
    device = 'cpu'
    entanglement_processor = EntanglementProcessor(n_qubits=2, device=device)
    superposition_processor = SuperpositionProcessor(device=device)

    # Create Bell State and calculate its entanglement entropy
    bell_state = entanglement_processor.create_bell_state()
    print("Bell State:", bell_state)

    entropy = entanglement_processor.get_entanglement_entropy(bell_state, partition=1)
    print("Entanglement Entropy of Bell State:", entropy)

    # Create Equal Superposition and apply phase shift
    equal_superposition = superposition_processor.create_equal_superposition(n_qubits=2)
    print("Equal Superposition State:", equal_superposition)

    phase_shifted_superposition = superposition_processor.apply_phase_shift(equal_superposition, angle=np.pi / 4)
    print("Phase-Shifted Superposition:", phase_shifted_superposition)
```

```python
# Main implementation fixes
class QuantumStateTest(unittest.TestCase):
    def setUp(self):
        self.device = 'cpu'
        self.dtype = torch.float64
        self.tolerance = 1e-6

    def _partial_trace(self, state: ComplexTensor, trace_out: int = 1) -> torch.Tensor:
        """
        Compute reduced density matrix by tracing out specified qubit.
        trace_out: which qubit to trace out (0 or 1)
        """
        # Reshape to 2x2x2x2 to represent 2-qubit state
        state_matrix = self._density_matrix(state)
        state_tensor = state_matrix.reshape(2, 2, 2, 2)

        if trace_out == 1:
            # Trace out second qubit
            reduced = torch.einsum('ijik->jk', state_tensor)
        else:
            # Trace out first qubit
            reduced = torch.einsum('ijkj->ik', state_tensor)

        return reduced

    def _calculate_entropy(self, state: ComplexTensor) -> float:
        """Calculate von Neumann entropy of reduced density matrix"""
        # Get reduced density matrix
        reduced_density = self._partial_trace(state)

        # Calculate eigenvalues
        eigenvals = torch.linalg.eigvalsh(reduced_density)

        # Filter small eigenvalues and calculate entropy
        eigenvals = eigenvals[eigenvals > self.tolerance]
        entropy = -torch.sum(eigenvals * torch.log2(eigenvals + self.tolerance)).real.item()

        return entropy

    def _density_matrix(self, state: ComplexTensor) -> torch.Tensor:
        """Convert state vector to density matrix"""
        # Convert to complex tensor
        state_complex = state.forward()
        # Reshape to column vector
        state_vec = state_complex.view(-1, 1)
        # Compute outer product
        return torch.matmul(state_vec, state_vec.conj().t())

    def test_bell_state(self):
        """Test Bell state preparation and properties"""
        # Create Bell state |Φ⁺⟩ = (|00⟩ + |11⟩)/√2
        real = torch.tensor([1/np.sqrt(2), 0, 0, 1/np.sqrt(2)], dtype=self.dtype)
        imag = torch.zeros(4, dtype=self.dtype)
        bell_state = ComplexTensor(real, imag)

        # Test normalization
        norm = bell_state.abs().pow(2).sum()
        self.assertAlmostEqual(norm.item(), 1.0, delta=self.tolerance)

        # Visualize Bell state
        self._plot_state(bell_state, "Bell State")

        # Calculate and verify entropy
        entropy = self._calculate_entropy(bell_state)
        print(f"Bell State Entropy: {entropy:.6f}")
        self.assertAlmostEqual(entropy, 1.0, delta=self.tolerance)

        # Verify maximally entangled state probabilities
        probs = bell_state.abs().pow(2)
        self.assertAlmostEqual(probs[0].item(), 0.5, delta=self.tolerance)
        self.assertAlmostEqual(probs[3].item(), 0.5, delta=self.tolerance)
        self.assertAlmostEqual(probs[1].item(), 0.0, delta=self.tolerance)
        self.assertAlmostEqual(probs[2].item(), 0.0, delta=self.tolerance)
```

```python
# Main implementation fixes
class QuantumStateTest(unittest.TestCase):
    def setUp(self):
        self.device = 'cpu'
        self.dtype = torch.float64
        self.tolerance = 1e-6

    def _partial_trace(self, state: ComplexTensor, trace_out: int = 1) -> torch.Tensor:
        """
        Compute reduced density matrix by tracing out specified qubit.
        trace_out: which qubit to trace out (0 or 1)
        """
        # Reshape to 2x2x2x2 to represent 2-qubit state
        state_matrix = self._density_matrix(state)
        state_tensor = state_matrix.reshape(2, 2, 2, 2)

        if trace_out == 1:
            # Trace out second qubit
            reduced = torch.einsum('ijik->jk', state_tensor)
        else:
            # Trace out first qubit
            reduced = torch.einsum('ijkj->ik', state_tensor)

        return reduced

    def _calculate_entropy(self, state: ComplexTensor) -> float:
        """Calculate von Neumann entropy of reduced density matrix"""
        # Get reduced density matrix
        reduced_density = self._partial_trace(state)

        # Calculate eigenvalues
        eigenvals = torch.linalg.eigvalsh(reduced_density)

        # Filter small eigenvalues and calculate entropy
        eigenvals = eigenvals[eigenvals > self.tolerance]
        entropy = -torch.sum(eigenvals * torch.log2(eigenvals + self.tolerance)).real.item()

        return entropy

    def _density_matrix(self, state: ComplexTensor) -> torch.Tensor:
        """Convert state vector to density matrix"""
        # Convert to complex tensor
        state_complex = state.forward()
        # Reshape to column vector
        state_vec = state_complex.view(-1, 1)
        # Compute outer product
        return torch.matmul(state_vec, state_vec.conj().t())

    def test_bell_state(self):
        """Test Bell state preparation and properties"""
        # Create Bell state |Φ⁺⟩ = (|00⟩ + |11⟩)/√2
        real = torch.tensor([1/np.sqrt(2), 0, 0, 1/np.sqrt(2)], dtype=self.dtype)
        imag = torch.zeros(4, dtype=self.dtype)
        bell_state = ComplexTensor(real, imag)

        # Test normalization
        norm = bell_state.abs().pow(2).sum()
        self.assertAlmostEqual(norm.item(), 1.0, delta=self.tolerance)

        # Visualize Bell state
        self._plot_state(bell_state, "Bell State")

        # Calculate and verify entropy
        entropy = self._calculate_entropy(bell_state)
        print(f"Bell State Entropy: {entropy:.6f}")
        self.assertAlmostEqual(entropy, 1.0, delta=self.tolerance)

        # Verify maximally entangled state probabilities
        probs = bell_state.abs().pow(2)
        self.assertAlmostEqual(probs[0].item(), 0.5, delta=self.tolerance)
        self.assertAlmostEqual(probs[3].item(), 0.5, delta=self.tolerance)
        self.assertAlmostEqual(probs[1].item(), 0.0, delta=self.tolerance)
        self.assertAlmostEqual(probs[2].item(), 0.0, delta=self.tolerance)
```

```python
import torch
import numpy as np
from typing import Dict, List, Tuple, Optional, Union
from complextensor import ComplexTensor
import logging
from dataclasses import dataclass
from contextlib import contextmanager

# Configure logging with more detail
logging.basicConfig(level=logging.DEBUG,
                   format='%(asctime)s - %(name)s - %(levelname)s - %(message)s')
logger = logging.getLogger(__name__)

@dataclass
class ShapeError(Exception):
    expected: tuple
    actual: tuple
    message: str = "Shape mismatch"

    def __str__(self):
        return f"{self.message}: expected {self.expected}, got {self.actual}"

@contextmanager
def shape_validation():
    try:
        yield
    except RuntimeError as e:
        if "size mismatch" in str(e):
            logger.error(f"Shape validation failed: {str(e)}")
            raise ShapeError(expected="unknown", actual="unknown",
                           message=str(e))
        raise

class QuantumStateAnalyzer:
    def __init__(self, n_qubits: int, device: str = 'cpu'):
        self.n_qubits = n_qubits
        self.hilbert_dim = 2 ** n_qubits
        self.device = device
        self.dtype = torch.complex128  # Use complex128 for better precision

        logger.debug(f"Initialized QuantumStateAnalyzer with {n_qubits} qubits "
                    f"on device {device}, Hilbert space dimension: {self.hilbert_dim}")

        # Validate device
        if not torch.cuda.is_available() and device == 'cuda':
            logger.warning("CUDA not available, falling back to CPU")
            self.device = 'cpu'

    def _validate_state_shape(self, state: ComplexTensor, expected_dim: int) -> None:
        """Validate state shape matches expected dimensions"""
        actual_shape = state.shape
        expected_shape = (expected_dim,)

        if actual_shape != expected_shape:
            raise ShapeError(expected=expected_shape, actual=actual_shape,
                           message="Invalid state shape")

    def _validate_kraus_op(self, operator: torch.Tensor, expected_shape: Tuple[int, int]) -> None:
        """Validate Kraus operator shape and properties"""
        if operator.shape != expected_shape:
            raise ShapeError(expected=expected_shape, actual=operator.shape,
                           message="Invalid Kraus operator shape")

    def _to_complex(self, tensor: torch.Tensor) -> torch.Tensor:
        """Convert real tensor to complex"""
        if not tensor.is_complex():
            return tensor.to(dtype=self.dtype, device=self.device)
        return tensor.to(device=self.device)

    def _apply_amplitude_damping(self, state: ComplexTensor, gamma: float) -> ComplexTensor:
        """Apply amplitude damping channel with shape validation and proper dtype handling"""
        logger.debug(f"Applying amplitude damping with gamma={gamma}")

        try:
            # Create Kraus operators with proper dtype
            kraus_0 = torch.tensor([[1., 0.],
                                  [0., np.sqrt(1-gamma)]],
                                 device=self.device,
                                 dtype=self.dtype)

            kraus_1 = torch.tensor([[0., np.sqrt(gamma)],
                                  [0., 0.]],
                                 device=self.device,
                                 dtype=self.dtype)

            # Validate Kraus operators
            self._validate_kraus_op(kraus_0, (2, 2))
            self._validate_kraus_op(kraus_1, (2, 2))

            # Get complex state tensor
            state_complex = state.forward()
            logger.debug(f"State shape before reshape: {state_complex.shape}")

            # Reshape with validation
            with shape_validation():
                state_matrix = state_complex.reshape(2, -1)
                logger.debug(f"State matrix shape: {state_matrix.shape}")

                # Apply Kraus operators
                result = torch.mm(kraus_0, state_matrix) + torch.mm(kraus_1, state_matrix)
                logger.debug(f"Result shape before final reshape: {result.shape}")

                result = result.reshape(-1)
                logger.debug(f"Final result shape: {result.shape}")

            return ComplexTensor(result.real, result.imag)

        except Exception as e:
            logger.error(f"Error in amplitude damping: {str(e)}")
            raise

    def analyze_decoherence(self, state: ComplexTensor, time_points: torch.Tensor) -> Dict:
        """Analyze decoherence with error handling and validation"""
        logger.info("Starting decoherence analysis")

        try:
            # Validate inputs
            self._validate_state_shape(state, self.hilbert_dim)
            if not torch.is_tensor(time_points):
                time_points = torch.tensor(time_points, device=self.device, dtype=torch.float64)

            coherence_times = []
            purities = []

            for t in time_points:
                logger.debug(f"Processing time point t={t.item()}")
                gamma = 1 - torch.exp(-t/20.0)

                try:
                    decohered = self._apply_amplitude_damping(state, gamma.item())
                    coherence = self._calculate_coherence(decohered)
                    purity = self._calculate_purity(decohered)

                    coherence_times.append(coherence)
                    purities.append(purity)

                except Exception as e:
                    logger.error(f"Error processing time point {t}: {str(e)}")
                    raise

            decoherence_time = self._estimate_decoherence_time(coherence_times, time_points)

            results = {
                'coherence_evolution': torch.tensor(coherence_times, device=self.device),
                'purity_evolution': torch.tensor(purities, device=self.device),
                'decoherence_time': decoherence_time,
                'parameters': {
                    'n_qubits': self.n_qubits,
                    'hilbert_dim': self.hilbert_dim,
                    'time_points': len(time_points)
                }
            }

            logger.info("Decoherence analysis completed successfully")
            return results

        except Exception as e:
            logger.error(f"Error in decoherence analysis: {str(e)}")
            raise

    def quantum_debug_info(self, state: ComplexTensor) -> Dict:
        """Generate debug information for quantum state"""
        return {
            'shape': state.shape,
            'device': state.device,
            'dtype': state.dtype,
            'norm': state.abs().pow(2).sum().item(),
            'requires_grad': state.real.requires_grad,
            'is_complex': state.forward().is_complex()
        }

# Usage example with error handling
def analyze_quantum_states():
    try:
        n_qubits = 2
        analyzer = QuantumStateAnalyzer(n_qubits)

        # Create states with debug info
        bell_state = analyzer.create_bell_state()
        logger.info(f"Bell state debug info: {analyzer.quantum_debug_info(bell_state)}")

        time_points = torch.linspace(0, 100, 1000, device=analyzer.device)
        results = analyzer.analyze_decoherence(bell_state, time_points)

        return results

    except Exception as e:
        logger.error(f"Analysis failed: {str(e)}")
        raise

if __name__ == "__main__":
    try:
        results = analyze_quantum_states()
        print("Analysis Results:", results)
    except Exception as e:
        logger.error(f"Main execution failed: {str(e)}")
        raise
```

```python
# New cell for quantum state creation methods
def create_bell_state(self) -> ComplexTensor:
    """
    Create a Bell state |Φ⁺⟩ = (|00⟩ + |11⟩)/√2
    """
    logger.debug("Creating Bell state")
    try:
        # Initialize state vector
        state = torch.zeros(4, dtype=torch.float64, device=self.device)
        state[0] = state[3] = 1/np.sqrt(2)

        logger.debug(f"Bell state shape: {state.shape}, dtype: {state.dtype}")

        # Validate state
        if not torch.allclose(torch.sum(state**2), torch.tensor(1.0)):
            raise ValueError("Bell state not normalized")

        return ComplexTensor(
            state,
            torch.zeros_like(state, device=self.device)
        )

    except Exception as e:
        logger.error(f"Failed to create Bell state: {str(e)}")
        raise

def create_ghz_state(self) -> ComplexTensor:
    """
    Create a GHZ state |GHZ⟩ = (|000...0⟩ + |111...1⟩)/√2
    """
    logger.debug(f"Creating GHZ state for {self.n_qubits} qubits")
    try:
        # Initialize state vector
        state = torch.zeros(self.hilbert_dim, dtype=torch.float64, device=self.device)
        state[0] = state[-1] = 1/np.sqrt(2)

        logger.debug(f"GHZ state shape: {state.shape}, dtype: {state.dtype}")

        return ComplexTensor(
            state,
            torch.zeros_like(state, device=self.device)
        )

    except Exception as e:
        logger.error(f"Failed to create GHZ state: {str(e)}")
        raise

def create_w_state(self) -> ComplexTensor:
    """
    Create a W state |W⟩ = (|100...0⟩ + |010...0⟩ + ... + |000...1⟩)/√n
    """
    logger.debug(f"Creating W state for {self.n_qubits} qubits")
    try:
        # Initialize state vector
        state = torch.zeros(self.hilbert_dim, dtype=torch.float64, device=self.device)

        # Set single excitation states
        normalization = 1.0/np.sqrt(self.n_qubits)
        for i in range(self.n_qubits):
            state[2**i] = normalization

        logger.debug(f"W state shape: {state.shape}, dtype: {state.dtype}")

        return ComplexTensor(
            state,
            torch.zeros_like(state, device=self.device)
        )

    except Exception as e:
        logger.error(f"Failed to create W state: {str(e)}")
        raise

# Add methods to QuantumStateAnalyzer
QuantumStateAnalyzer.create_bell_state = create_bell_state
QuantumStateAnalyzer.create_ghz_state = create_ghz_state
QuantumStateAnalyzer.create_w_state = create_w_state

# Test state creation
if __name__ == "__main__":
    try:
        analyzer = QuantumStateAnalyzer(n_qubits=2)

        # Create and validate states
        bell = analyzer.create_bell_state()
        ghz = analyzer.create_ghz_state()
        w = analyzer.create_w_state()

        print("Bell state:", bell)
        print("GHZ state:", ghz)
        print("W state:", w)

    except Exception as e:
        logger.error(f"State creation test failed: {str(e)}")
        raise
```

```python
# Fixed state creation methods with consistent dtype handling
def create_bell_state(self) -> ComplexTensor:
    """
    Create a Bell state |Φ⁺⟩ = (|00⟩ + |11⟩)/√2
    """
    logger.debug("Creating Bell state")
    try:
        # Initialize state vector with explicit dtype
        norm_factor = torch.tensor(1/np.sqrt(2), dtype=torch.float32, device=self.device)
        state = torch.zeros(4, dtype=torch.float32, device=self.device)
        state[0] = state[3] = norm_factor

        logger.debug(f"Bell state shape: {state.shape}, dtype: {state.dtype}")

        # Validate state with matching dtype
        norm = torch.sum(state**2)
        target = torch.tensor(1.0, dtype=torch.float32, device=self.device)

        if not torch.allclose(norm, target, rtol=1e-5):
            raise ValueError(f"Bell state not normalized: {norm.item()} != 1.0")

        return ComplexTensor(
            state,
            torch.zeros_like(state)
        )

    except Exception as e:
        logger.error(f"Failed to create Bell state: {str(e)}")
        raise

def create_ghz_state(self) -> ComplexTensor:
    """
    Create a GHZ state |GHZ⟩ = (|000...0⟩ + |111...1⟩)/√2
    """
    logger.debug(f"Creating GHZ state for {self.n_qubits} qubits")
    try:
        # Initialize with explicit dtype
        norm_factor = torch.tensor(1/np.sqrt(2), dtype=torch.float32, device=self.device)
        state = torch.zeros(self.hilbert_dim, dtype=torch.float32, device=self.device)
        state[0] = state[-1] = norm_factor

        logger.debug(f"GHZ state shape: {state.shape}, dtype: {state.dtype}")

        # Validate normalization
        norm = torch.sum(state**2)
        target = torch.tensor(1.0, dtype=torch.float32, device=self.device)

        if not torch.allclose(norm, target, rtol=1e-5):
            raise ValueError(f"GHZ state not normalized: {norm.item()} != 1.0")

        return ComplexTensor(
            state,
            torch.zeros_like(state)
        )

    except Exception as e:
        logger.error(f"Failed to create GHZ state: {str(e)}")
        raise

def create_w_state(self) -> ComplexTensor:
    """
    Create a W state |W⟩ = (|100...0⟩ + |010...0⟩ + ... + |000...1⟩)/√n
    """
    logger.debug(f"Creating W state for {self.n_qubits} qubits")
    try:
        # Initialize with explicit dtype
        normalization = torch.tensor(1.0/np.sqrt(self.n_qubits),
                                   dtype=torch.float32, device=self.device)
        state = torch.zeros(self.hilbert_dim, dtype=torch.float32, device=self.device)

        # Set single excitation states
        for i in range(self.n_qubits):
            state[2**i] = normalization

        logger.debug(f"W state shape: {state.shape}, dtype: {state.dtype}")

        # Validate normalization
        norm = torch.sum(state**2)
        target = torch.tensor(1.0, dtype=torch.float32, device=self.device)

        if not torch.allclose(norm, target, rtol=1e-5):
            raise ValueError(f"W state not normalized: {norm.item()} != 1.0")

        return ComplexTensor(
            state,
            torch.zeros_like(state)
        )

    except Exception as e:
        logger.error(f"Failed to create W state: {str(e)}")
        raise

def validate_quantum_state(self, state: ComplexTensor, name: str = "state") -> None:
    """Validate quantum state properties"""
    try:
        # Check normalization
        norm = state.abs().pow(2).sum()
        target = torch.tensor(1.0, dtype=state.real.dtype, device=state.device)

        if not torch.allclose(norm, target, rtol=1e-5):
            raise ValueError(f"{name} not normalized: {norm.item()} != 1.0")

        # Check shape
        expected_dim = self.hilbert_dim
        if state.shape != (expected_dim,):
            raise ValueError(f"{name} has wrong shape: {state.shape} != ({expected_dim},)")

        logger.debug(f"Validated {name}: shape={state.shape}, norm={norm.item():.6f}")

    except Exception as e:
        logger.error(f"State validation failed for {name}: {str(e)}")
        raise

# Update QuantumStateAnalyzer class dtype
class QuantumStateAnalyzer:
    def __init__(self, n_qubits: int, device: str = 'cpu'):
        self.n_qubits = n_qubits
        self.hilbert_dim = 2 ** n_qubits
        self.device = device
        self.dtype = torch.float32  # Changed to float32

        logger.debug(f"Initialized QuantumStateAnalyzer with {n_qubits} qubits "
                    f"on device {device}, dtype={self.dtype}")

# Add methods to QuantumStateAnalyzer
QuantumStateAnalyzer.create_bell_state = create_bell_state
QuantumStateAnalyzer.create_ghz_state = create_ghz_state
QuantumStateAnalyzer.create_w_state = create_w_state
QuantumStateAnalyzer.validate_quantum_state = validate_quantum_state

# Test state creation with validation
if __name__ == "__main__":
    try:
        analyzer = QuantumStateAnalyzer(n_qubits=2)

        # Create and validate states
        bell = analyzer.create_bell_state()
        analyzer.validate_quantum_state(bell, "Bell state")

        ghz = analyzer.create_ghz_state()
        analyzer.validate_quantum_state(ghz, "GHZ state")

        w = analyzer.create_w_state()
        analyzer.validate_quantum_state(w, "W state")

        print("\nQuantum State Properties:")
        for name, state in [("Bell", bell), ("GHZ", ghz), ("W", w)]:
            print(f"\n{name} state:")
            print(f"Real part: {state.real}")
            print(f"Imaginary part: {state.imag}")
            print(f"Norm: {state.abs().pow(2).sum().item():.6f}")

    except Exception as e:
        logger.error(f"State creation test failed: {str(e)}")
        raise
```
